```python
#changing the datatype to float from str
df_bse["10-Year Return(%)"] = pd.to_numeric(df_bse["10-Year Return(%)"], downcast="signed")

#filling the blank values with nan
df_bse['10-Year Return(%)'] = df_bse['10-Year Return(%)'].fillna(0)

# # finding the differnece in in bse500 in new column delta

df_bse['delta']=(df_bse['52 Week High'].sub(df_bse['Price'], axis = 0).div(df_bse['52 Week High']))
```

```python


# define a lambda function to count positive and negative numbers
count_pos_neg = lambda x: pd.Series([sum(x > 0), sum(x < 0)], index=['Positive', 'Negative'])


# group by 'Group' column and apply count_pos_neg function
result = industry.groupby('Industry')['3-Year Return'].apply(count_pos_neg).reset_index()

print(result)
```

```python
result=pd.DataFrame(result)
result=result.sort_values(by='3-Year Return',ascending=False).head(15)
result
```

```python


plt.figure(figsize=(25,20))
plt.bar(result['Industry'],result['3-Year Return'] )

plt.title('Counts of Positive and Negative Numbers by Group')


plt.xlabel('Industry')
plt.ylabel('3-Year Return')
plt.show()
```
